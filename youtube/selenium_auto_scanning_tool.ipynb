```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import pandas as pd
import time

query = "samsung+kg+mdm+unlock"
# links = []


driver = webdriver.Edge()
driver.get(f"https://www.youtube.com/results?search_query={query}")
```

```python
# filter by "This month" option:

driver.find_element(By.ID, "filter-button").click()

time.sleep(2)

filter = driver.find_elements(By.ID, "endpoint")
for i in filter:
   if(i.text == "This month"):
      i.click()

time.sleep(2)
```

```python
# scroll page
i = 10
body = driver.find_element(By.TAG_NAME, 'body') 
while(i > 0):
    k = 5
    while(k >= 0):
        body.send_keys(Keys.PAGE_DOWN)  # Scroll down
        k -= 1
    
    time.sleep(1)
    i -= 1

```

```python
# elements = []
def ignore_case(element):
    ignore_word = ["T Mobile", "US Cellular", "Sprint USA", "Unlock Service", "Xfinity USA", "Cricket USA", "FRP", "Boost USA", "Verizon USA", "Spectrum", "Lost mode", "Huawei",
                   "Xiaomi", "screen lock", "TFN", "iphone"]
    text = element.text.lower()
    for i in ignore_word:
        if i.lower() in text:
            return True
    else:
        return False
    

```

```python
# get elements:
data = {'titles' : [],
        'links' : [],
        'dates' : [],
        'contents' : []}
elements = []
els = driver.find_elements(By.ID, "video-title")

for c, e in enumerate(els):
    if(e.text != "" and ignore_case(e) != True):
        elements.append(e)
        # print(f"{c} : {e.text}")
```

```python
# save to arrays
titles = []
links = []
dates = []
ytlinks = []
descriptions = []

# elem = driver.find_elements(By.TAG_NAME, "a")
for i, e in enumerate(elements):
    link = e.get_attribute('href')
    if "shorts" in link: 
        continue
    links.append(link)
    # titles.append(e.text)
    print(f"{i} - {e.text} - {e.get_attribute('href')}")
```

```python
def get_title(titles):
    # print("get_title:")
    title = driver.find_elements(By.ID, "title")
    for t in title:
        # print(t.text)
        if(t.text != ""):
            titles.append(t.text)
            return None
```

```python

# get describe
def get_content(descriptions):
    # print("get_content:")
    time.sleep(2)
    driver.find_element(By.ID, "expand").click()
    description = driver.find_element(By.ID, "description-inline-expander")
    # print(description.text)
    descriptions.append(description.text)
```

```python
def get_date(dates):
    # print("get_date:")
    date = driver.find_elements(By.ID, "info")
    # title = driver.find_element(By.ID, "title")
    for i, a in enumerate(date):
        # print(f"{i} : {a.text}")
        if(a.text != ""):
            # titles.append(title.text)
            dates.append(a.text)
            return None
```

```python
def get_info(link, titles, dates, descriptions):
    driver.get(link)
    time.sleep(2)
    print(link)
    # dates.append(date)
    get_date(dates)
    get_content(descriptions)
    get_title(titles)
```

```python
early_stop = 3

count  = 0
for link in links:
    if(early_stop == 0):
        break
    # early_stop -= 1
    ytlinks.append(link)
    count += 1
    print(count)
    get_info(link, titles, dates, descriptions)
    #khong can time sleep
    time.sleep(4)
```
